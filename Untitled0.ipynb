```python
# firt tect preproceesing  on inpur columns
# import some libaray  for text preproccessing
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# list of stopwords
stop_words=stopwords.words("english")
print(stop_words)
```

```python
from types import new_class
# useeer definr fucntion
def clean_text(msg):# clean_text() user defined function
# firts covert into lower case(small )
    msg=msg.lower()
    # apply  Tokenizier ,split words from given sentence
    # tokenizer will not tAKE THE HELPING VARD  LIKE (IS ,THE ,AM ,ARE ETC)
    token = word_tokenize(msg)
    #print(token)
    # filtering  to remove sspecal character  and number
    new_token=[w for  w in token if w.isalpha()] #to remove special character and numbers
    #print(new_token)
    #next step to remove the  stopword
    new_token=[w for w in new_token  if w  not in stop_words ]
    # Next  step  apply Lemmitazation class
    # create the object  of WordNetLemmatizer class
    # it help to  make the same word  mulitple same words which hai same meaning butit is pressent in multiple form
    #   grammer  that will replace with one same word
    lemm=WordNetLemmatizer()
    new_token=[lemm.lemmatize(w) for w in new_token]
    m = ' '.join(new_token) # join is used to join the token  into full line  Because  the token used only  for preproccessing
    return m
```

```python
# call function
# example
#clean_text("My name is vishal study studes ,history historys")
```

```python
# call  Clean_text()  in message  input column  in datafrane df
df["Message"]=df["Message"].apply(clean_text)
```

```python
#   now  once  we have the  clean text  in hand ,the  next things  that we  would  to do
# is convert  the input  text into vector/number\
# first  use CountVectorizer
# call  inbuilt  class CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer
```
